```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

## Matplotlib - Functional Method

```python
plt.plot(x,y)
#plt.show()
```

```python
plt.plot(x,y,color='green')
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

## Multiplot

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

## Object Oriented Matplotlib

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

## Tuple unpacking method

```python
fig,axes= plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('second plot')

plt.tight_layout()
```

## Figure Size and DPI - Dots/Inch

```python
fig = plt.figure(figsize=(3,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,5))
axes[0].plot(x,y,'g')
axes[1].plot(y,x,'r')

plt.tight_layout()
```

## Add Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Square')
ax.plot(x,x**3,label = 'X cubed')

ax.legend()
#ax.legend(loc=(0.1,0.1)) #location code for your legends
```

## Line customization

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=10,alpha=0.5) #alpha is to make line more transparent
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=3,linestyle='--') #we can use ls instead of linestyle argument
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,ls='-.') 
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2,linestyle='steps') 
```

## Show limited plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=1,linestyle='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
