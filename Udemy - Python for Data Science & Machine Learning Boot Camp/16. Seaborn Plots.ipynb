```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Distribution Plots

```python
sns.distplot(tips['total_bill'])
```

### Removing kde - kernel density estimation

```python
sns.distplot(tips['total_bill'],kde=False)
```

### Add bins

```python
sns.distplot(tips['total_bill'],kde=False,bins=50)
```

## JointPlots - BiVariate(2 variables)

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

### Adding kind

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') #regression line
```

## Pair Plots

```python
#plots relation between numerical variables
sns.pairplot(tips)
```

### Hue - Adding a Categorical variable 

```python
sns.pairplot(tips,hue = 'sex')
```

#### Palettes for color

```python
sns.pairplot(tips,hue = 'sex',palette='husl')
```

## Rugplot

```python
sns.rugplot(tips['total_bill'])
```

## KDEPlot

```python
sns.kdeplot(tips['total_bill'])
```

## Categorical Plots

### Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

#### Using Estimator for mean/median/std/custom

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.median)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.mean)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### Countplot

```python
sns.countplot(x='sex',data=tips)
```

### Boxplot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

#### Hue - Add categorical variable

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

### Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

### Strip Plot

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
#Jitter - scatters dots a bit
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',jitter=True,split=True)
```

### Swarm Plot - combination of swarm plot and Violin Plot
- Drawback 
- - Does not scale properly when data point are large
- - Lot of computation power is required
- - Not recommended for large datasets

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

#### Combining violin plot and swarmplot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='white')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

## Grids

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

### Pair Grid

```python
sns.PairGrid(iris)
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

#### combine maps

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

### FacetGrid

```python
tips.head()
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.scatterplot,'total_bill','tip')
```
